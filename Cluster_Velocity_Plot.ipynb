```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Simulation of cluster equation with "x" amount of bacteria in cluster.
dp = 300 # Num. of data points
x = np.linspace(0, dp, dp)
y = [(1 - 0.95*((0.15+0.95)/(0.15+1)))/(1 + 0.95*((0.15+0.95)/(0.15+1)) + 2*0.15*((x*0.15)/(0.15+1)) ) for x in range(dp)]
y = np.array(y)
print(len(x))
print(len(y))

fig, axes = plt.subplots(figsize=(10, 10))
plot = axes.plot(x, y)
```
