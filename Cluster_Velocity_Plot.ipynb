```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objs as go
```

```python
# Simulation of cluster equation with "x" amount of bacteria in cluster.
dp = 300 # Num. of data points
x = np.linspace(0, dp, dp)
y = [(1 - 0.95*((0.15+0.95)/(0.15+1)))/(1 + 0.95*((0.15+0.95)/(0.15+1)) + 2*0.15*((x*0.15)/(0.15+1)) ) for x in range(dp)]
y = np.array(y)
print(len(x))
print(len(y))

fig, axes = plt.subplots(figsize=(10, 10))
plot = axes.plot(x, y)
```

```python
# Average run time with "x" amount of bacteria in cluster.
dp = 10 # Num. of data points
x = np.linspace(1, dp, (dp-1))
y = [( 1/(0.15+0.85)/(x*0.15) ) for x in range(1, dp)]
y = np.array(y)

fig, axes = plt.subplots(figsize=(10, 10))
plot = axes.plot(x, y)
```

```python
# Effective run time of cluster approximation
num_bac = 2
dp = 10 # Num. of data points
t = np.linspace(0, dp, dp)

# Probability of cluster still running function
def run_cluster(num_bac, dp):
    Q = [( np.exp(-(num_bac*t)/5.67) ) for t in range(dp)]
    return np.array(Q)

x_values = np.linspace(0, dp, dp)

# create traces
trace0 = go.Scatter(
    x = x_values,
    y = run_cluster(1, dp),
    mode = 'lines',
    name = 'Num_Bac = 1'
)
# The second trace
trace1 = go.Scatter(
    x = x_values,
    y = run_cluster(2, dp),
    mode = 'lines',
    name = 'Num_Bac = 2'
)
# create the data
data = [trace0, trace1]  # assign traces to data

# crate the fig
fig = go.Figure(data=data)
fig.show()
```
